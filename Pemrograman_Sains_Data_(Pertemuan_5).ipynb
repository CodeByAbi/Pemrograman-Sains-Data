**Horizantally Stacked Bar Graph**

```python
divisions = ["Div-A", "Div-B", "Div-C", "Div-D", "Div-E"]
divisions_average_marks = [70, 82, 73, 65, 68]
boys_averages_marks = [68,67,77,61,70]

index = np.arange(5)
width = 0.30

plt.bar(index, divisions_average_marks, width, color='green',label='Division Marks')
plt.bar(index+width, boys_averages_marks, width, color='blue', label='Boys Marks')
plt.title("Horizantally Stacked Bar Graph")

plt.ylabel("Marks")
plt.xlabel("Divisons")
plt.xticks(index+ width/2, divisions)

plt.legend(loc='best')
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
random = []
for _ in range(50):
  random.append(np.random.randint(1,100))
#create a dataframe format and save a random number inside
df = pd.DataFrame(random,columns =['Random'])
#create a single line cart
plt.figure(figsize = (20,5))
plt.plot(df["Random"], marker = 'o')
plt.title('Single Line Chart with Random Numbers')
plt.xticks(df.index)
plt.xlabel('Row')
plt.ylabel('Random Numbers')
for i_x, i_y in zip(df.index, df['Random']):
  plt.text(x = i_x, y = i_y, s = '({}, {})'.format(i_x,i_y), rotation = -45)
  plt.grid(True)

plt.show()
```

**Studi Kasus 2 :**

Menganalisis hubungan antara jumlah tidur malam seorang individu dan produktivitas kerja mereka di hari berikutnya. Data ini mencakup 20 data dengan informasi tentang jumlah tidur malam (dalam jam) dan produktivitas kerja (dalam skala 0-10).

```python
import matplotlib.pyplot as plt

# Data
cabang = ['Cabang A', 'Cabang B', 'Cabang C', 'Cabang D', 'Cabang E', 'Cabang F', 'Cabang G', 'Cabang H', 'Cabang I', 'Cabang J', 'Cabang K', 'Cabang L', 'Cabang M', 'Cabang N', 'Cabang O']
jumlah_produk_terjual = [50, 60, 45, 70, 55, 80, 40, 65, 75, 85, 35, 55, 90, 95, 30]
pendapatan = [1000, 1200, 900, 1400, 1100, 1600, 800, 1300, 1500, 1700, 700, 1100, 1800, 1900, 600]

# Plotting
fig, ax = plt.subplots(figsize=(10, 6))
bar_width = 0.35
index = range(len(cabang))
bar1 = ax.bar(index, jumlah_produk_terjual, bar_width, label='Jumlah Produk Terjual')
bar2 = ax.bar([i + bar_width for i in index], pendapatan, bar_width, label='Pendapatan')

ax.set_xlabel('Cabang')
ax.set_ylabel('Jumlah / Pendapatan (Dolar)')
ax.set_title('Penjualan Berdasarkan Cabang')
ax.set_xticks([i + bar_width / 2 for i in index])
ax.set_xticklabels(cabang, rotation=45, ha='right')
ax.legend()

plt.tight_layout()
plt.show()
```
